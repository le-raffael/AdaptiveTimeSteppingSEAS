```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10.00,7.00))
plt.rcParams.update({'font.size': 18})

x = np.linspace(-5,5,1000)
m = 3
h = 1
A = 0.04
t = 1
y = A * np.sqrt(t) * np.cos(m * x**2 / (h * t))

n = 50
x_simul = np.zeros(4*n)
y_simul = np.zeros(4*n)
for k in range(0,50):    
    x_simul[2*n - 2 * k - 1] = -np.sqrt(2./m * np.pi * k)
    y_simul[2*n - 2 * k - 1] = -A
    x_simul[2*n - 2 * k - 2] = -np.sqrt(np.pi/m + 2./m * np.pi * k)
    y_simul[2*n - 2 * k - 2] = A
    x_simul[2*n + 2 * k] = np.sqrt(2./m * np.pi * k)
    y_simul[2*n + 2 * k] = -A
    x_simul[2*n + 2 * k+1] = np.sqrt(np.pi/m + 2./m * np.pi * k)
    y_simul[2*n + 2 * k+1] = A

x_simul[2*n - 2 * k - 1] = -np.sqrt(2./m * np.pi * (n-k))
y_simul[2*n - 2 * k - 1] = -A

plt.plot(x,y,'k',label='Feynman kernel')
plt.plot(x_simul[2:-1],y_simul[2:-1],'--k',label='Photon')

plt.xlabel('x')
plt.ylabel('Real part of the photon amplitude')

plt.xlim([-3,3])
plt.legend(bbox_to_anchor=(0.8, -0.1), ncol = 2)

plt.savefig('freeParticleResults.png', format='png',dpi=300, bbox_inches='tight')

plt.show()

```
