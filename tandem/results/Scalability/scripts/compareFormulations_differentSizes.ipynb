```python
import numpy as np
import sys
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cbook as cbook
import matplotlib.ticker as mticker
import csv
import pylab
from scipy.interpolate import interp1d


np.set_printoptions(threshold=sys.maxsize)
```

```python
time_list_size52 = []
dt_array_size52 = []
V_array_size52 = []
F_array_size52 = []
rhs_count_array_size52 = []
ksp_count_array_size52 = []
max_S_size52 = []
min_S_size52 = []
max_PSI_size52 = []
min_PSI_size52 = []
time_old = 0.0



#don't plot first two components
first = 1
second = 1

with open('../results/TANDEM_size52_400years__AS_extendedode_5dp_Vrtol1e-08__EQ_extendedode_5dp_Vrtol1e-07.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if second:
            second = 0
        if first:
            first = 0
            second = 1
        else:
            time = row['time']
            dt = float(time) - time_old
            time_old = float(time)
            time_list_size52.append(time)
            dt_array_size52.append(dt)

            V_array_size52.append(row['Vmax'])
            F_array_size52.append(row['fmax'])
            rhs_count_array_size52.append(row['count_rhs'])
            ksp_count_array_size52.append(row['KSP_iteration'])
            max_S_size52.append(row['maxS'])
            min_S_size52.append(row['minS'])
            max_PSI_size52.append(row['maxPSI'])
            min_PSI_size52.append(row['minPSI'])
            
            
            if (row['t_init']):
                initialization_time_size52 = row['t_init']
            if (row['t_solve']):
                solve_time_size52 = row['t_solve']
            if (row['t_total']):
                total_time_size52 = row['t_total']

time_array_size52 = np.asarray(time_list_size52, dtype=np.float64)
dt_array_size52 = np.asarray(dt_array_size52, dtype=np.float64)
V_array_size52 = np.asarray(V_array_size52, dtype=np.float64)
F_array_size52 = np.asarray(F_array_size52, dtype=np.float64)
rhs_count_array_size52 = np.asarray(rhs_count_array_size52, dtype=np.float64)
ksp_count_array_size52 = np.asarray(ksp_count_array_size52, dtype=np.float64)
max_S_size52 = np.asarray(max_S_size52, dtype=np.float64)
min_S_size52 = np.asarray(min_S_size52, dtype=np.float64)
max_PSI_size52 = np.asarray(max_PSI_size52, dtype=np.float64)
min_PSI_size52 = np.asarray(min_PSI_size52, dtype=np.float64)

time_years_size52 = time_array_size52 / (365.2425*24*3600)
```
