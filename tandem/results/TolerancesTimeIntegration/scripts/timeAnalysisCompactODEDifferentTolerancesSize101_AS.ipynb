```python
sigma = 1e-7
C = 1
e_S = 4e-12
e_psi = 2e-14
e_V = 2e-24
```

```python
def dgdpsi():

    b = 0.015
    V0 = 1e-6
    L = 0.008
    f = 0.6
    psi_max = 0.7
    
    return V0 / L * (np.exp((f - psi_max) / b))

def dgdV():

    b = 0.015
    V0 = 1e-6
    L = 0.008
    f = 0.6
    psi_max = 0.7
    
    return b / L
        
```

```python
# compact ODE
t_S = sigma * C * e_S + e_S + e_V
t_psi = sigma * C * e_psi + e_psi + dgdpsi() * e_psi + dgdV() * e_V
print("Newton tolerances: \n\t -tolerance in S: ", t_S, "\n\t -tolerance in psi:", t_psi)
```
