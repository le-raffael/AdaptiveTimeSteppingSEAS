```python
time_list_BDF_52 = []
dt_array_BDF_52 = []
V_array_BDF_52 = []
rhs_count_array_BDF_52 = []
time_old = 0.0

#don't plot first two components
first = 1
second = 1

with open('../results/TANDEM_size52__AS_compactode_BDF0_Stol1e-08__EQ_compactode_BDF0_Stol1e-08.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if second:
            second = 0
        if first:
            first = 0
            second = 1
        else:
            time = row['time']
            dt = float(time) - time_old
            time_old = float(time)
            time_list_BDF_52.append(time)
            dt_array_BDF_52.append(dt)

            V_array_BDF_52.append(row['Vmax'])
            rhs_count_array_BDF_52.append(row['count_rhs'])

time_array_BDF_52 = np.asarray(time_list_BDF_52, dtype=np.float64)
dt_array_BDF_52 = np.asarray(dt_array_BDF_52, dtype=np.float64)
V_array_BDF_52 = np.asarray(V_array_BDF_52, dtype=np.float64)
rhs_count_array_BDF_52 = np.asarray(rhs_count_array_BDF_52, dtype=np.float64)

time_years_BDF_52 = time_array_BDF_52 / (365.2425*24*3600)
```
