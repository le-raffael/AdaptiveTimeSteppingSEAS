```python
import numpy as np
import sys
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cbook as cbook
import csv
import pylab
from scipy.interpolate import interp1d


np.set_printoptions(threshold=sys.maxsize)
```

```python
V = np.logspace(-11,1,1000,base=10)

eps  = 1e-12
s    = 50       # = sigma
n    = 2.670 * 32.038120320 / 2.0
Vmax = 10

eps_V = V * eps / (s + V * n)

a = (Vmax / (n * (Vmax*Vmax / 2. + Vmax*s / n)))**2 * s
b = (1 + a*s) / n - 2. / n * np.sqrt(s * a)

a = eps * a
b = eps * b

a = eps / s
b = 0

approx = a * V + b

plt.figure(figsize=(10.00,7.00))
plt.rcParams.update({'font.size': 18})
plt.plot(V, eps_V,'k',label='exact lower bound')
plt.plot(V, approx,'--k',label='affine approximate')
plt.xlabel(r'slip rate $V$ [m/s]')
plt.ylabel(r'error $\varepsilon^V$ [m/s]')
plt.yscale('log')
plt.xscale('log')
plt.legend()

plt.savefig('../results/ErrorRelationPSI_VAbsoluteRelative.png', format='png',dpi=300, bbox_inches='tight')
plt.show()
```
