```python
import numpy as np
import sys
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cbook as cbook
import csv
import pylab
from scipy.interpolate import interp1d


np.set_printoptions(threshold=sys.maxsize)
```

```python
def absoluteError(e_V):
    amin = 0.01
    V0 = 1e-6
    return amin * np.log(1 + 1 / (np.arcsinh(0.5)*np.sqrt(5)*V0) * e_V)

def relativeError(e_V):
    amin = 0.01
    psimax = 0.85
    Vmax = 5.0
    V0 = 1e-6
    return amin * np.log(1 + 1 / (np.arcsinh(Vmax / (2*V0))) * e_V)
```

```python
order_abs = np.linspace(-15,2, 100)
order_rel = np.linspace(-15,2, 100)

eps_V_abs = 10**order_abs
eps_V_rel = 10**order_rel

print(absoluteError(1e-10))
print(relativeError(1e-7))

eps_psi_abs = absoluteError(eps_V_abs)
eps_psi_rel = relativeError(eps_V_rel)


plt.figure(figsize=(10.00,7.00))
plt.rcParams.update({'font.size': 18})
plt.plot(eps_V_rel, eps_psi_rel,'-.k',label="relative error")
plt.plot(eps_V_abs, eps_psi_abs,'-k',label="absolute error")
plt.xlabel("error in $V$")
plt.ylabel("error in $\psi$")
plt.yscale('log')
plt.xscale('log')
plt.legend()
plt.savefig('../results/ErrorRelationPSI_VAbsoluteRelative.png', format='png',dpi=300, bbox_inches='tight')
plt.show()
```
