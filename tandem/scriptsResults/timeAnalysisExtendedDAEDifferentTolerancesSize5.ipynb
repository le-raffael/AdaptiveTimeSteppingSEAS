```python
time_list_tolem1 = []
dt_array_tolem1 = []
V_array_tolem1 = []
F_array_tolem1 = []
rhs_count_array_tolem1 = []
time_old = 0.0

#don't plot first two components
first = 1
second = 1

with open('../results/TANDEM_size5__AS_extendeddae_BDF0_Stol1e-01__EQ_extendeddae_BDF0_Stol1e-01.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if second:
            second = 0
        if first:
            first = 0
            second = 1
        else:
            time = row['time']
            dt = float(time) - time_old
            time_old = float(time)
            time_list_tolem1.append(time)
            dt_array_tolem1.append(dt)

            V_array_tolem1.append(row['Vmax'])
            F_array_tolem1.append(row['fmax'])
            rhs_count_array_tolem1.append(row['count_rhs'])

time_array_tolem1 = np.asarray(time_list_tolem1, dtype=np.float64)
dt_array_tolem1 = np.asarray(dt_array_tolem1, dtype=np.float64)
V_array_tolem1 = np.asarray(V_array_tolem1, dtype=np.float64)
F_array_tolem1 = np.asarray(F_array_tolem1, dtype=np.float64)
rhs_count_array_tolem1 = np.asarray(rhs_count_array_tolem1, dtype=np.float64)

time_years_tolem1 = time_array_tolem1 / (365.2425*24*3600)
```
