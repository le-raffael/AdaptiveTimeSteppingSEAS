```python
time_list_BDF_40 = []
dt_array_BDF_40 = []
V_array_BDF_40 = []
max_dfdpsi_array_BDF_40 = []
rhs_count_array_BDF_40 = []
time_old = 0.0

#don't plot first two components
first = 1
second = 1

with open('../results/TANDEM_size40__AS_extendeddae_BDF0_sor-gmres_Stol1e-06__EQ_extendeddae_BDF0_sor-gmres_Stol1e-06.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        if second:
            second = 0
        if first:
            first = 0
            second = 1
        else:
            time = row['time']
            dt = float(time) - time_old
            time_old = float(time)
            time_list_BDF_40.append(time)
            dt_array_BDF_40.append(dt)

            V_array_BDF_40.append(row['Vmax'])
            max_dfdpsi_array_BDF_40.append(row['maxErrorFactorPSI'])
            rhs_count_array_BDF_40.append(row['count_rhs'])

time_array_BDF_40 = np.asarray(time_list_BDF_40, dtype=np.float64)
dt_array_BDF_40 = np.asarray(dt_array_BDF_40, dtype=np.float64)
V_array_BDF_40 = np.asarray(V_array_BDF_40, dtype=np.float64)
rhs_count_array_BDF_40 = np.asarray(rhs_count_array_BDF_40, dtype=np.float64)
max_dfdpsi_array_BDF_40 = np.asarray(max_dfdpsi_array_BDF_40, dtype=np.float64)

time_years_BDF_40 = time_array_BDF_40 / (365.2425*24*3600)
```
