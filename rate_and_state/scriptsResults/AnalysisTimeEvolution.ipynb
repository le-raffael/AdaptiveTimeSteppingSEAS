```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cbook as cbook
import csv
```

```python
time_array_RKF45 = []
dt_array_RKF45 = []
V_array_RKF45 = []
psi_array_RKF45 = []
V_err_array_RKF45 = []
psi_err_array_RKF45 = []
time_old = 0.0

with open('../results/result_RKF45.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        time = row['time']
        dt = float(time) - time_old
        time_old = float(time)
        time_array_RKF45.append(time)
        dt_array_RKF45.append(dt)
        
        V_array_RKF45.append(row['V'])
        psi_array_RKF45.append(row['psi'])
        V_err_array_RKF45.append(row['V_err'])
        psi_err_array_RKF45.append(row['psi_err'])
        
time_array_RKF45 = np.asarray(time_array_RKF45, dtype=np.float32)
V_array_RKF45 = np.asarray(V_array_RKF45, dtype=np.float32)
psi_array_RKF45 = np.asarray(psi_array_RKF45, dtype=np.float32)
V_err_array_RKF45 = np.asarray(V_err_array_RKF45, dtype=np.float32)
psi_err_array_RKF45 = np.asarray(psi_err_array_RKF45, dtype=np.float32)
```

```python
time_array_BDF12 = []
dt_array_BDF12 = []
V_array_BDF12 = []
psi_array_BDF12 = []
V_err_array_BDF12 = []
psi_err_array_BDF12 = []
time_old = 0.0

with open('../results/result_BDF12.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        time = row['time']
        dt = float(time) - time_old
        time_old = float(time)
        time_array_BDF12.append(time)
        dt_array_BDF12.append(dt)

        V_array_BDF12.append(row['V'])
        psi_array_BDF12.append(row['psi'])
        V_err_array_BDF12.append(row['V_err'])
        psi_err_array_BDF12.append(row['psi_err'])
        
time_array_BDF12 = np.asarray(time_array_BDF12, dtype=np.float32)
V_array_BDF12 = np.asarray(V_array_BDF12, dtype=np.float32)
psi_array_BDF12 = np.asarray(psi_array_BDF12, dtype=np.float32)
V_err_array_BDF12 = np.asarray(V_err_array_BDF12, dtype=np.float32)
psi_err_array_BDF12 = np.asarray(psi_err_array_BDF12, dtype=np.float32)
```

```python
time_array_BDF23 = []
dt_array_BDF23 = []
V_array_BDF23 = []
psi_array_BDF23 = []
V_err_array_BDF23 = []
psi_err_array_BDF23 = []
time_old = 0.0

with open('../results/result_BDF23.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        time = row['time']
        dt = float(time) - time_old
        time_old = float(time)
        time_array_BDF23.append(time)
        dt_array_BDF23.append(dt)

        V_array_BDF23.append(row['V'])
        psi_array_BDF23.append(row['psi'])
        V_err_array_BDF23.append(row['V_err'])
        psi_err_array_BDF23.append(row['psi_err'])
        
time_array_BDF23 = np.asarray(time_array_BDF23, dtype=np.float32)
V_array_BDF23 = np.asarray(V_array_BDF23, dtype=np.float32)
psi_array_BDF23 = np.asarray(psi_array_BDF23, dtype=np.float32)
V_err_array_BDF23 = np.asarray(V_err_array_BDF23, dtype=np.float32)
psi_err_array_BDF23 = np.asarray(psi_err_array_BDF23, dtype=np.float32)
```

```python
plt.figure(figsize=(10.00,7.00))
plt.rcParams.update({'font.size': 18})
params = {'text.latex.preamble' : [r'\usepackage{siunitx}', r'\usepackage{amsmath}']}
plt.rcParams.update(params)

fig, ax1 = plt.subplots(figsize=(10.00,7.00))

ax2 = ax1.twinx()


ax2.plot(time_array_BDF12, V_array_BDF12,'g',linewidth=8, label = r'$V$ - BDF1')
ax1.plot(time_array_BDF12, psi_array_BDF12,'--g',linewidth=8, label = r'$\bar{\psi}$ - BDF1')

ax2.plot(time_array_BDF23, V_array_BDF23,'r',linewidth=4, label = r'$V$ - BDF2')
ax1.plot(time_array_BDF23, psi_array_BDF23,'--r',linewidth=4, label = r'$\bar{\psi}$ - BDF2')

ax2.plot(time_array_RKF45, V_array_RKF45,'b',linewidth=2, label = r'$V$ - RKF4')
ax1.plot(time_array_RKF45, psi_array_RKF45,'--b',linewidth=2, label = r'$\bar{\psi}$ - RKF4')


ax2.set_xlabel("time")
ax2.set_ylabel(r'$V$')
ax1.set_ylabel(r'$\bar{\psi}$')

ax1.legend(loc='center left')
ax2.legend(loc='center right')

plt.savefig('../results/timeEvolutionValues.png', format='png',dpi=300, bbox_inches='tight')

plt.show()
```

```python
plt.figure(figsize=(10.00,7.00))
plt.rcParams.update({'font.size': 18})

plt.plot(time_array_BDF12, dt_array_BDF12,'-g', label = "BDF1")
plt.plot(time_array_BDF23, dt_array_BDF23,'-r', label = "BDF2")
plt.plot(time_array_RKF45, dt_array_RKF45,'-b', label = "RKF4")

plt.xlabel("time")
plt.ylabel(r'$h$')

plt.yscale('log')

plt.ylim(1e-3, 1)

plt.legend()

plt.savefig('../results/timeEvolutionDT.png', format='png',dpi=300, bbox_inches='tight')

plt.show()
```

```python
plt.figure(figsize=(10.00,7.00))
plt.rcParams.update({'font.size': 18})

plt.plot(time_array_BDF12, psi_err_array_BDF12,'-g', label = "BDF1")
plt.plot(time_array_BDF23, psi_err_array_BDF23,'-r', label = "BDF2")
plt.plot(time_array_RKF45, psi_err_array_RKF45,'-b', label = "RKF4")

plt.xlabel("time")
plt.ylabel(r'error in $\bar{\psi}$')

plt.yscale('log')
plt.ylim([1e-9, 1e-4])

plt.legend()

plt.savefig('../results/timeEvolutionPSIerror.png', format='png',dpi=300, bbox_inches='tight')

plt.show()
```

```python
plt.figure(figsize=(10.00,7.00))
plt.rcParams.update({'font.size': 18})

plt.plot(time_array_BDF12, V_err_array_BDF12,'-g', label = "BDF1")
plt.plot(time_array_BDF23, V_err_array_BDF23,'-r', label = "BDF2")
plt.plot(time_array_RKF45, V_err_array_RKF45,'-b', label = "RKF4")

plt.xlabel("time")
plt.ylabel(r'error in $V$')

plt.yscale('log')
plt.ylim([1e-9, 1e-1])

plt.legend()

plt.savefig('../results/timeEvolutionVerror.png', format='png',dpi=300, bbox_inches='tight')

plt.show()
```
