```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cbook as cbook
import csv
```

```python
time_array_RKF45 = []
estimated_err_array_RKF45 = []
real_err_array_RKF45 = []

with open('../results/errorEstimate_RKF45.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        time_array_RKF45.append(row['time'])
        estimated_err_array_RKF45.append(row['local_err'])
        real_err_array_RKF45.append(row['err_expected'])
        
time_array_RKF45 = np.asarray(time_array_RKF45, dtype=np.float32)
estimated_err_array_RKF45 = np.asarray(estimated_err_array_RKF45, dtype=np.float32)
real_err_array_RKF45 = np.asarray(real_err_array_RKF45, dtype=np.float32)
```

```python
time_array_BDF12 = []
estimated_err_array_BDF12 = []
estimated_Lagrange_err_array_BDF12 = []
real_err_array_BDF12 = []

with open('../results/errorEstimate_BDF12.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        time_array_BDF12.append(row['time'])
        estimated_err_array_BDF12.append(row['local_err'])
        estimated_Lagrange_err_array_BDF12.append(row['err_Lagrange'])
        real_err_array_BDF12.append(row['err_expected'])
        
time_array_BDF12 = np.asarray(time_array_BDF12, dtype=np.float32)
estimated_err_array_BDF12 = np.asarray(estimated_err_array_BDF12, dtype=np.float32)
estimated_Lagrange_err_array_BDF12 = np.asarray(estimated_Lagrange_err_array_BDF12, dtype=np.float32)
real_err_array_BDF12 = np.asarray(real_err_array_BDF12, dtype=np.float32)
```

```python
time_array_BDF23 = []
estimated_err_array_BDF23 = []
estimated_Lagrange_err_array_BDF23 = []
real_err_array_BDF23 = []

with open('../results/errorEstimate_BDF23.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        time_array_BDF23.append(row['time'])
        estimated_err_array_BDF23.append(row['local_err'])
        estimated_Lagrange_err_array_BDF23.append(row['err_Lagrange'])
        real_err_array_BDF23.append(row['err_expected'])
        
time_array_BDF23 = np.asarray(time_array_BDF23, dtype=np.float32)
estimated_err_array_BDF23 = np.asarray(estimated_err_array_BDF23, dtype=np.float32)
estimated_Lagrange_err_array_BDF23 = np.asarray(estimated_Lagrange_err_array_BDF23, dtype=np.float32)
real_err_array_BDF23 = np.asarray(real_err_array_BDF23, dtype=np.float32)
```

```python
plt.figure(figsize=(10.00,7.00))
plt.rcParams.update({'font.size': 27})

plt.plot(time_array_RKF45, estimated_err_array_RKF45,'bx',label = "estimated error")
plt.plot(time_array_RKF45, real_err_array_RKF45,'b', label = "real error")

plt.xlabel("time")
plt.ylabel("local truncation error")

plt.yscale('log')

plt.legend()

plt.savefig('../results/errorEstimateRKF45.png', format='png',dpi=300)

plt.show()
```

```python
plt.figure(figsize=(10.00,7.00))
plt.rcParams.update({'font.size': 27})

plt.plot(time_array_BDF12, estimated_err_array_BDF12,'gx',label = "estimated error")
plt.plot(time_array_BDF12, estimated_Lagrange_err_array_BDF12,'go',label = "estimated error - Lagrange")
plt.plot(time_array_BDF12, real_err_array_BDF12,'g', label = "real error")

plt.xlabel("time")
plt.ylabel("local truncation error")

plt.yscale('log')

plt.legend()

plt.savefig('../results/errorEstimateBDF12.png', format='png',dpi=300)

plt.show()
```
