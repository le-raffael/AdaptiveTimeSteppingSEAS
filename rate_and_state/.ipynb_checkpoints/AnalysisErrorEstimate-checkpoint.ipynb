```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cbook as cbook
import csv
```

```python
time_array_RKF45 = []
estimated_err_array_RKF45 = []
real_err_array_RKF45 = []

with open('build/errorEstimate_RKF45.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        time_array_RKF45.append(row['time'])
        estimated_err_array_RKF45.append(row['local_err'])
        real_err_array_RKF45.append(row['err_expected'])
        
time_array_RKF45 = np.asarray(time_array_RKF45, dtype=np.float32)
estimated_err_array_RKF45 = np.asarray(estimated_err_array_RKF45, dtype=np.float32)
real_err_array_RKF45 = np.asarray(real_err_array_RKF45, dtype=np.float32)
```

```python
time_array_DP = []
estimated_err_array_DP = []
real_err_array_DP = []

with open('build/errorEstimate_RKDP.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        time_array_DP.append(row['time'])
        estimated_err_array_DP.append(row['local_err'])
        real_err_array_DP.append(row['err_expected'])
        
time_array_DP = np.asarray(time_array_DP, dtype=np.float32)
estimated_err_array_DP = np.asarray(estimated_err_array_DP, dtype=np.float32)
real_err_array_DP = np.asarray(real_err_array_DP, dtype=np.float32)
```

```python
time_array_BE = []
estimated_err_array_BE = []
real_err_array_BE = []

with open('build/errorEstimate_BE.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        time_array_BE.append(row['time'])
        estimated_err_array_BE.append(row['local_err'])
        real_err_array_BE.append(row['err_expected'])
        
time_array_BE = np.asarray(time_array_BE, dtype=np.float32)
estimated_err_array_BE = np.asarray(estimated_err_array_BE, dtype=np.float32)
real_err_array_BE = np.asarray(real_err_array_BE, dtype=np.float32)
```

```python
plt.figure(figsize=(10.00,10.00))
#plt.plot(time_array_RKF45, estimated_err_array_RKF45,'gx',label = "estimated error (RKF)")
#plt.plot(time_array_RKF45, real_err_array_RKF45,'g', label = "real error (RKF)")
#plt.plot(time_array_DP, estimated_err_array_DP,'mx',label = "estimated error (RKDP)")
#plt.plot(time_array_DP, real_err_array_DP,'m', label = "real error (RKDP)")
plt.plot(time_array_BE, estimated_err_array_BE,'bx',label = "estimated error (BE)")
plt.plot(time_array_BE, real_err_array_BE,'b', label = "real error (BE)")

plt.xlabel("time")
plt.ylabel("local truncation error")

plt.legend()

plt.show()
```
